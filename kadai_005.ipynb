```python
import matplotlib.pyplot as plt
import numpy as np
import japanize_matplotlib

fig, axes = plt.subplots(2, 3, tight_layout=True)

# 折れ線グラフ
y0 = [1, 2, -5, 2]
axes[0, 0].set_title('折れ線グラフ')
axes[0, 0].plot(y0)

# sin関数
x0 = np.linspace(0, 10, 150)
x1 = 2 + 2 * np.sin(2 * x0)
axes[0, 1].set_title('sin関数')
axes[0, 1].plot(x0, x1, linewidth=2.0)

# ヒストグラム
x2 = np.random.normal(15, 5, 2000)
axes[0, 2].set_title('ヒストグラム')
axes[0, 2].hist(x2)

# 箱ひげ図
x3 = np.random.normal(300, 100, 3000)
axes[1, 0].set_title('箱ひげ図')
axes[1, 0].boxplot(x3)

# 散布図
np.random.seed(3)
x4 = 4 + np.random.rand(2, 60)
y4 = 4 + np.random.rand(2, 60)
axes[1, 1].set_title('散布図')
axes[1, 1].scatter(x4[0], y4[0])
axes[1, 1].scatter(x4[1], y4[1])

# 円グラフ
x5 = np.array([30, 20, 60])
label = ['豹', '虎', 'チベットスナギツネ']
axes[1, 2].set_title('好きな動物')
axes[1, 2].pie(x5, labels=label)

plt.show()
```
