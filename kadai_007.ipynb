```python
import matplotlib.pyplot as plt
import numpy as np
import japanize_matplotlib
import pandas as pd

df = pd.read_csv("sample_pandas_6.csv")
category_df = pd.read_csv('category.csv')
df = pd.merge(df, category_df[['商品番号', 'カテゴリー']], how='inner', on='商品番号')
# df
# 出現頻度をカウントし、棒グラフ作成
df['カテゴリー'].value_counts().plot(kind='bar')
plt.title("カテゴリー別")
plt.xlabel("カテゴリー")
plt.ylabel("個数")
plt.show()

df.groupby('商品番号')['注文数'].count()
df.groupby('商品番号')['注文数'].describe().plot(kind='bar' )
plt.title("商品別")
plt.xlabel("商品番号")
plt.ylabel("注文数")
plt.show()
```
