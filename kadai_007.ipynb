```python
import matplotlib.pyplot as plt
import numpy as np
import japanize_matplotlib
import pandas as pd

df = pd.read_csv("sample_pandas_6.csv")
category_df = pd.read_csv('category.csv')
df = pd.merge(df, category_df[['', '']], how='inner', on='')
# df
# 
df[''].value_counts().plot(kind='bar')
plt.title("")
plt.xlabel("")
plt.ylabel("")
plt.show()

df.groupby('')[''].count()
df.groupby('')[''].describe().plot(kind='bar' )
plt.title("")
plt.xlabel("")
plt.ylabel("")
plt.show()
```
